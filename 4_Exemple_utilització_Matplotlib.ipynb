# 4.2 Importar dades

Normalment s'hauria d'instal·lar manualment Matplotlib i Numpy, però a Google Colab ja ve preinstalada la llibreria.

## En aquest quadern es servir la llibreria NUMPY

## Per pujar l'arxiu i importar les llibreries s'ha d'executar el següent bloc de codi:

````python
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files
archivo = files.upload()
archivo = list(archivo.keys())[0]
data = np.genfromtxt(archivo, delimiter=';', dtype=None, encoding=None ,skip_header=6, skip_footer=2, names=[ 'Edat','Homes', 'Dones', 'Total'])
edat = np.arange(101)
````

# 4.3 Visualització gràfica

## Per fer les gràfiques:

````python
fig, axs = plt.subplots(3, 1)
fig.set_size_inches(15,15)

axs[0].plot(edat, data['Homes'])
axs[0].set_title('Homes')
axs[0].grid()
axs[1].plot(edat, data['Dones'], 'tab:green')
axs[1].set_title('Dones')
axs[1].grid()
axs[2].plot(edat, data['Total'], 'tab:red')
axs[2].set_title('Total')
axs[2].grid()

for ax in axs.flat:
    ax.set(xlabel='Edat', ylabel='Nº de morts')

for ax in axs.flat:
    ax.label_outer()
````

## Per veure els 3 gràfics junts:

````python
fig = plt.figure()
fig.set_size_inches(15,10)
ax1 = fig.add_subplot(111)

ax1.plot(edat, data['Homes'], color='r', label='the data')
ax1.plot(edat, data['Dones'], color='g', label='the data')
ax1.plot(edat, data['Total'], color='b', label='the data')
fig.suptitle('Blau = Total\nVermell = Homes\nVerd = Dones', fontsize=16)
ax1.set(xlabel='Edat', ylabel='Nº Morts')
plt.grid()
````
