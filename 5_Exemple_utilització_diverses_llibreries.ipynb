# 1.2 Utilització de llibreries

```python
#comandaments per carregar les llibreries
import numpy as np
import plotly.graph_objs as go
import pandas as pd
#amb la instrucció pd.read_excel es carregen desde el document pujat les dades mitjançant la llibreria de pandas
fijo=pd.read_excel('/content/Datos.xlsx',sheet_name='Telefono fijo')
movil=pd.read_excel('/content/Datos.xlsx',sheet_name='telefono movil')
internet=pd.read_excel('/content/Datos.xlsx',sheet_name='internet')
banda=pd.read_excel('/content/Datos.xlsx',sheet_name='banda ancha')
ordenador=pd.read_excel('/content/Datos.xlsx',sheet_name='ordenador')
```

```python
from matplotlib.axis import XAxis
#go.Scatter es la instrucció per crear la gràfica, que asociem a una variable
Total=go.Scatter(x=fijo['Año'],y=fijo['Total nacional'],name='Total nacional')
Rioja=go.Scatter(x=fijo['Año'],y=fijo['Rioja (La)'],name='Rioja')
Andalucia=go.Scatter(x=fijo['Año'],y=fijo['Andalucía'],name='Andalucia')
Aragon=go.Scatter(x=fijo['Año'],y=fijo['Aragón'],name='Aragó')
Asturias=go.Scatter(x=fijo['Año'],y=fijo['Asturias (Principado de)'],name='Asturias')
Balears=go.Scatter(x=fijo['Año'],y=fijo['Balears (Illes)'],name='Balears')
Canarias=go.Scatter(x=fijo['Año'],y=fijo['Canarias'],name='Canarias')
Cantabria=go.Scatter(x=fijo['Año'],y=fijo['Cantabria'],name='Cantabria')
Castilla=go.Scatter(x=fijo['Año'],y=fijo['Castilla y León'],name='Castilla i Lleó')
Cataluña=go.Scatter(x=fijo['Año'],y=fijo['Cataluña'],name='Catalunya')
ComunitatValenciana=go.Scatter(x=fijo['Año'],y=fijo['Comunitat Valenciana'],name='Comunitat Valenciana')
Extremadura=go.Scatter(x=fijo['Año'],y=fijo['Extremadura'],name='Extremadura')
Galicia=go.Scatter(x=fijo['Año'],y=fijo['Galicia'],name='Galicia')
Madrid=go.Scatter(x=fijo['Año'],y=fijo['Madrid (Comunidad de)'],name='Comunitat de Madrid')
Murcia=go.Scatter(x=fijo['Año'],y=fijo['Murcia (Región de)'],name='Murcia')
Navarra=go.Scatter(x=fijo['Año'],y=fijo['Navarra (Comunidad Foral de)'],name='Navarra')
PaisVasco=go.Scatter(x=fijo['Año'],y=fijo['País Vasco'],name='País Basc')
Ceuta=go.Scatter(x=fijo['Año'],y=fijo['Ceuta'],name='Ceuta')
Melilla=go.Scatter(x=fijo['Año'],y=fijo['Melilla'],name='Melilla')
#adjuntem totes les variables que corresponen a les gràfiques de punts
data=[Total,Andalucia,Aragon,Asturias,Balears,Canarias,Cantabria,Castilla,Cataluña,ComunitatValenciana,
      Extremadura,Galicia,Madrid,Murcia,Navarra,PaisVasco,Rioja,Ceuta,Melilla]
#amb la instrucció go.Layout afegim les característiques a la gràfica
layout=go.Layout(title='Possessió de telèfon fix',xaxis=dict(title='Any'),yaxis=dict(title='Percentatge'))
#go.Figure representem totes les variables de manera conjunta amb les característiques definides anteriorment
fig=go.Figure(data=data,layout=layout)
#aquesta instrucció es per aconseguir el gràfic en pantalla
fig.show(renderer="colab")
```

```python
#per realitzar el gràfic de barres de manera sencilla es necessaria carregar una altra llibreria
import plotly.express as px
#Es carreguen en una variable les dades per aquesta representació 
inter=pd.read_excel('/content/Datos.xlsx',sheet_name='movilfijo')
#amb px.bar() es crea la grafica de barres que s'asocia a una variable
inter=px.bar(inter,x='Año',y=['Telefono movil','Telefono fijo'],title='Telefono fijo vs Telefono movil',barmode='overlay')
#amb la instrucció seguent es canvia les característiques
inter.update_layout(title = "Telefono fijo vs Telefono movil", 
      xaxis_title = 'Año', yaxis_title =  
      'Porcentaje',legend_title_text=None)
inter.show(render='colab')
```
