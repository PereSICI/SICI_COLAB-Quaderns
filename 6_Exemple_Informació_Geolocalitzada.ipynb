S'inicialitza:
1. El tipus de cartografia (robin)
2. La ressolució
3. La distribució de línees de costa

```python
eq_map =Basemap(projection='robin',
                lon_0=0, 
                resolution="h",
                area_thresh=1000.0,
                llcrnrlon=-136.25,
                llcrnrlat=56,
                urcrnrlon=-134.25,
                urcrnrlat=57.75
                )
```

Es dibuixen les línies de costa i païssos

```python
eq_map.drawcoastlines()
eq_map.drawcountries()
```

Es defineix el color dels païssos

```python
eq_map.fillcontinents(color = "#e6e6ff")
eq_map.drawmapboundary()
```

Es dibuixen meridians i paral·lels

```python
eq_map.drawmeridians(np.arange(0,360,30))
eq_map.drawparallels(np.arange(-90,90,30))
```
