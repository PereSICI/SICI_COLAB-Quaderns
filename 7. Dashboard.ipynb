# 1.4 Detall tècnic

La llibreria Dash de Plotly permet la generació d'una aplicació en format html.
La idea principal es que es poden tenir les dades en un repositori, o bé en el propi quadern Colab. Alesjhores, el que cal es vincular el cuadern de Google Colab amb la seva interfície web.
D'aquesta forma, qualsevol usuari connectat a xarxa internet, pot navegar i accedir a la interfície web creada.
L'empresa Anvil explica, per exemple, com fer aquesta vinculació amb el seu producte comercial. Creat per Ryan, A URL:(https://anvil.works/learn/tutorials/google-colab-to-web-app) i a l'enllaç (https://colab.research.google.com/drive/17Y70UHd_dNtY5Dyv9OIJlUXVYjW5b5qZ?usp=sharing)
